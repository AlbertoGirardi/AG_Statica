# Doppio pendolo fisico

Il problema del doppio pendolo è un problema interessante perchè è un esempio di problema nonlineare e caotico.
Approcciamo la scrittura delle equazione della dinamica dal punto di vista energetico.
Definiamo come $T$ l'energia cinetica totale del sistema:
$$
T = \frac 1 2 J_{G_1}\dot \theta_1^2  + \frac 1 2 J_{G_2}\dot \theta_2^2 + \frac 1 2 m_{1}v_1^2 + \frac 1 2 m_{2}v_2^2
$$

mentre l'energia potenziale $V$ si scrive come:

$$
V = m_1gy_1 + m_2gy_2 
$$

Prima di tutto calcoliamo le seguenti quantità cinematiche:

**Posizione baricentro**
$$
\begin{align*}
x_1  &= \frac{l_1}{2}\sin(\theta_1)\\
y_1  &= - \frac{l_1}{2}\cos(\theta_1)\\
\end{align*}
$$

$$
\begin{align*}
x_2  &= l_1\sin(\theta_1) + \frac{l_2}{2}\sin(\theta_2) \\
y_2  &= - l_1\cos(\theta_1) -\frac{l_2}{2}\cos(\theta_2)
\end{align*}
$$
**Velocità baricentro**
$$
\begin{align*}
\dot x_1  = \frac{l_1}{2}\cos(\theta_1)\dot \theta_1\\
\dot y_1  = \frac{l_1}{2}\sin(\theta_1)\dot \theta_1
\end{align*}
$$
$$
\begin{align*}
\dot x_2  = l_1\cos(\theta_1)\dot \theta_1 + \frac{l_2}{2}\cos(\theta_2)\dot \theta_2 \\
\dot y_2  = +l_1\sin(\theta_1)\dot \theta_1 +\frac{l_2}{2}\sin(\theta_2)\dot \theta_2
\end{align*}
$$
Calcoliamo le velocità
$$
\begin{align*}
v_1^2&=  \dot x_1^2  + \dot y_1^2 \\
     &= \frac{l_1^2}{4} \cos^2(\theta_1) \dot \theta_1^2 + \frac{l_1^2}{4}\sin^2(\theta_1)\dot \theta_1^2 \\
     &= \frac{l_1^2}{4}\dot \theta_1^2
\end{align*}
$$
$$
\begin{align*}
v_2^2&=  \dot x_2^2  + \dot y_2^2 \\
     &= l_1^2 \cos^2(\theta_1) \dot \theta_1^2 + \frac{l_2^2}{4} \cos^2(\theta_2) \dot \theta_2^2+ l_1l_2 \cos(\theta_1)\cos(\theta_2) + l_1^2\sin^2(\theta_1)\dot \theta_1^2   + \frac{l_2^2}{4}\sin^2(\theta_2)\dot \theta_2^2  + l_1l_2 \sin(\theta_1)\sin(\theta_2) \\
     &= l_1^2\dot \theta_1^2\ + \frac{l_2^2}{4}\dot \theta_2^2 +  l_1l_2\cos(\theta_1 - \theta_2)\dot \theta_1 \dot \theta_2
\end{align*}
$$
Componiamo quindi l'energia cinetica:

$$
T  = \frac 1 2 J_{G_1}\dot \theta_1^2  + \frac 1 2 J_{G_2}\dot \theta_2^2 + \frac 1 2 m_{1} \frac{l_1^2}{4}\dot \theta_1^2  + \frac 1 2 m_{2}\bigl( l_1^2 \dot \theta_1^2\ + \frac{l_2^2}{4}\dot \theta_2^2 +  l_1l_2\cos(\theta_1 - \theta_2)\dot \theta_1 \dot \theta_2\bigl)
$$

Ora raccogliendo $\dot \theta_1$ e $\dot \theta_2$:

$$
\begin{align*}
T  &=& \frac 1 2 \bigl ( J_{G_1} +  m_{1}\frac{l_1^2}{4}\bigl) \dot \theta_1^2 + \frac 1 2 \bigl ( J_{G_2} +  m_{2}\frac{l_2^2}{4}\bigl) \dot \theta_2^2 +  \frac 1 2 m_{2} l_1^2 \dot \theta_1^2\  +  \frac 1 2m_2 l_1l_2\cos(\theta_1 - \theta_2)\dot \theta_1 \dot \theta_2 \\
   &=& \frac 1 2 J_{O_1} \dot \theta_1^2 + \frac 1 2 J_{O_2}  \dot \theta_2^2 +  \frac 1 2 m_{2} l_1^2 \dot \theta_1^2\  +  \frac 1 2m_2 l_1l_2\cos(\theta_1 - \theta_2)\dot \theta_1 \dot \theta_2
\end{align*}
$$

Dove abbiamo utilizzato il teorema del trasporto dei momenti di inezia dai baricentri alle cerniere.

Per l'energia potenziale si ha

$$
V = - m_1g\frac{l_1}{ 2} \cos(\theta_1) - m_2g\bigl(l_1\cos(\theta_1) + \frac{l_2}{ 2} \cos(\theta_2)\bigl) 
$$

Ora componiamo la lagrangiana del problema $L = T-V$. 

$$
L   =   \frac 1 2 J_{O_1} \dot \theta_1^2 + \frac 1 2 J_{O_2}  \dot \theta_2^2 +  \frac 1 2 m_{2} l_1^2 \dot \theta_1^2\  +  \frac 1 2 m_2 l_1l_2\cos(\theta_1 - \theta_2)\dot \theta_1 \dot \theta_2 + \bigl(\frac{m_1}{ 2} + m_2\bigl)gl_1\cos(\theta_1) + \frac {m_2}{2}g l_2 \cos(\theta_2) 
$$

Costruiamo i momenti generalizzati facendo 
$$
\begin{align*}
p_{\dot \theta_1}  &=& \frac{\partial L}{\partial \dot \theta_1} &=& \bigl(\underbrace{ J_{O_1} +   m_{2} l_1^2 }_{A_1} \bigl) \dot \theta_1 &+ \underbrace{\frac 1 2  m_2 l_1l_2\cos(\theta_1 - \theta_2)}_B\dot \theta_2  &= A_1\dot \theta_1  + B\dot \theta_2 \\
p_{\dot \theta_2}  &=& \frac{\partial L}{\partial \dot \theta_2} &=&  \underbrace{J_{O_2}}_{A_2} \dot \theta_2  &+ \underbrace{\frac 1 2  m_2 l_1l_2\cos(\theta_1 - \theta_2)}_B\dot \theta_1  &= A_2\dot \theta_2  + B\dot \theta_1\\
\end{align*}
$$

Da cui si ricava:
$$
\begin{align*}
\dot \theta_1  &=& \frac{A_2 p_{\dot \theta_1} -B p_{\dot \theta_2}}{A_1 A_2- B^2} \\
\dot \theta_2  &=& \frac{A_1 p_{\dot \theta_2} -B p_{\dot \theta_1}}{A_1 A_2- B^2}
\end{align*}
$$

Sappiamo inoltre che 
$$
\frac{d}{dt} \frac{\partial L}{\partial \dot \theta_1} = \dot p_{\dot \theta_1}= \frac{\partial L}{\partial \theta_1}\qquad
\frac{d}{dt} \frac{\partial L}{\partial \dot \theta_2} = \dot p_{\dot \theta_2}= \frac{\partial L}{\partial \theta_2} 
$$

Deriviamo infine come si scrivono le derivate dei momenti generalizzati:

$$
\begin{align*}
\dot p_{\dot \theta_1} &=& -\underbrace{\frac 1 2 m_2 l_1l_2\sin(\theta_1 - \theta_2)}_{D} \dot\theta_1 \dot\theta_2 &- \underbrace{(\frac {m_1}{2}+m_2)gl_1}_{C_1} \sin(\theta_1) & = -D \dot \theta_1 \dot\theta_2 - C_1\sin(\theta_1)\\
\dot p_{\dot \theta_2} &=& +\underbrace{ \frac 1 2 m_2 l_1l_2\sin(\theta_1 - \theta_2)}_{D} \dot\theta_1 \dot\theta_2 &- \underbrace{\frac {m_2}{2}gl_2}_{C_2}\sin(\theta_2) & = + D \dot \theta_1 \dot\theta_2 - C_2\sin(\theta_2)
\end{align*}
$$

L'integrazione numerica procederà costruendo il vettore di stato con le seguenti grandezze:
$$
\overline{u}=
\begin{bmatrix} 
\theta_1\\
\theta_2\\
p_{\dot{\theta_1}}\\
p_{\dot{\theta_2}}
\end{bmatrix} \quad \dot{\overline{u}} =
\begin{bmatrix} 
\dot{\theta_1}\\
\dot{\theta_2}\\
\dot{p}_{\dot{\theta_1}}\\
\dot{p}_{\dot{\theta_2}}
\end{bmatrix}
$$

```python
## https://scipython.com/blog/the-double-compound-pendulum/
## https://jakevdp.github.io/blog/2012/08/18/matplotlib-animation-tutorial/

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from   scipy.integrate import solve_ivp

class double_compound_pendulum:
    def __init__(self,  g,l1,l2,m1,m2,JG1,JG2):
        
        self.g   = g
        self.l1 = l1
        self.l2 = l2
        self.m1 = m1
        self.m2 = m2
        self.JG1 = JG1
        self.JG2 = JG2
 
    
    def __call__(self, t, u):
        theta1,theta2,pdtheta1,pdtheta2 = u
        g = self.g
        l1 = self.l1 
        l2 = self.l2
        m1 = self.m1
        m2 = self.m2 
        JG1 =self.JG1 
        JG2 =self.JG2 
        
        JO1 = (JG1 + m1*(l1/2.)**2)
        JO2 = (JG2 + m2*(l2/2.)**2)
      
        ct1t2 = np.cos(theta1 - theta2)
        st1t2 = np.sin(theta1 - theta2)
        
        A1 = JO1 + m2*l1**2.
        A2 = JO2
        B  = 0.5*m2*l1*l2*ct1t2

        C1 = (0.5*m1+m2)*g*l1*np.sin(theta1) 
        C2 = 0.5*m2*g*l2*np.sin(theta2) 
        D  = 0.5*m2*l1*l2*st1t2
        
        dtheta1   = (A2*pdtheta1- B*pdtheta2)/(A1*A2-B**2)
        dtheta2   = (A1*pdtheta2- B*pdtheta1)/(A1*A2-B**2)
        
        dpdtheta1 = -D*dtheta1*dtheta2-C1 
        dpdtheta2 = +D*dtheta1*dtheta2-C2  

        return [dtheta1, dtheta2,dpdtheta1,dpdtheta2]

def Hamiltonian(theta1 ,theta2, pdtheta1,pdtheta2, g,l1,l2,m1,m2,JG1,JG2):
        
    JO1 = (JG1 + m1*(l1/2.)**2)
    JO2 = (JG2 + m2*(l2/2.)**2)
      
    ct1t2 = np.cos(theta1 - theta2)
    st1t2 = np.sin(theta1 - theta2)
        
    A1 = JO1 + m2*l1**2.
    A2 = JO2
    B  = 0.5*m2*l1*l2*ct1t2

    C1 = (0.5*m1+m2)*g*l1*np.sin(theta1) 
    C2 = 0.5*m2*g*l2*np.sin(theta2) 
    D  = 0.5*m2*l1*l2*st1t2
        
    dtheta1   = (A2*pdtheta1- B*pdtheta2)/(A1*A2-B**2)
    dtheta2   = (A1*pdtheta2- B*pdtheta1)/(A1*A2-B**2)

    T =  0.5*JO1*dtheta1**2 + 0.5*JO2*dtheta2**2 + 0.5*m2*l1*dtheta1**2 + 0.5*l1*l2*ct1t2*dtheta1*dtheta2
    V = -0.5*m1*g*l1*np.cos(theta1) - m2*g*( l1*np.cos(theta1) + 0.5*l2*np.cos(theta2)) 

    H = T + V 

    return H


g = 9.81
l1=1.
l2=1.
m1=1.
m2=1.
JG1=1./12.*m1*l1**2
JG2=1./12.*m2*l2**2


T    = 10 #perido
N    = 1000 # numero di intervalli di tempo 
dt   = T/N # intervallo di tempo
solverstr = 'Radau' #Radau DOP853 RK45


t  = np.linspace(0,T,N+1)

theta10 = 120./180.*np.pi
theta20 = 180./180.*np.pi
pdtheta10 = 0. 
pdtheta20 = 0.
u0 = np.array([theta10,theta20,pdtheta10,pdtheta20])
# metodo Runge Kutta predictor corrector 4/5 ordine
sol = solve_ivp(double_compound_pendulum(g=g, l1=l1, l2=l2, m1=m1, m2=m2, JG1=JG1, JG2=JG2), [0, T], u0 , method = solverstr, dense_output=True)

# Unpack dynamical variables as a function of time.
theta1, theta2, pdtheta1, pdtheta2 = sol.sol(t)

H  = Hamiltonian(theta1 ,theta2, pdtheta1,pdtheta2, g,l1,l2,m1,m2,JG1,JG2)
H0 = Hamiltonian(theta10 ,theta20, pdtheta10,pdtheta20, g,l1,l2,m1,m2,JG1,JG2)
errdrift = abs((H-H0)/H0)
EDRIFT_TRESHOLD = 1e-3

if any(errdrift > EDRIFT_TRESHOLD ):
    print('Maximum energy drift exceeded')




# Convert to Cartesian coordinates of the two rods.
x0 =  np.zeros_like(theta1)
y0 =  np.zeros_like(theta1)
x1 =  l1 * np.sin(theta1)
y1 = -l1 * np.cos(theta1)
x2 =  x1 + l2 * np.sin(theta2)
y2 =  y1 - l2 * np.cos(theta2)

xmat = np.column_stack((x0,x1,x2))
ymat = np.column_stack((y0,y1,y2))

figs,axs = plt.subplots(1,2)  

axs[0].set_title(solverstr)
axs[0].axis([-l1-l2, l1+l2,-l1-l2, l1+l2])
axs[0].set_aspect('equal', adjustable='box')
line, = axs[0].plot(xmat[0,:], ymat[0,:], 'o-b', lw=2)


axs[1].set_title('drift error')
axs[1].axis([0, T,np.min(errdrift), np.max(errdrift) ])
line1, = axs[1].plot(t[0],errdrift[0],'-r')


def update(frame):
    line.set_data(xmat[frame,:],ymat[frame,:])
    line1.set_data(t[0:frame],errdrift[0:frame])
    return [line,line1]


ani = animation.FuncAnimation(fig=figs, func=update, frames=N+1, interval=dt*1000)
plt.show()
```
