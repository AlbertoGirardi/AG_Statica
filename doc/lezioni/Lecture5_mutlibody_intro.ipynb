# Cinematica e dinamica dei corpi rigidi vincolato in due dimensioni
## Introduzione


Nel caso di un corpo rigido vincolato oltre alle forze attive si devono tener in conto anche le forze di reazione dei vincoli sia interni che esterni. L'equazione vettoriale diventetà:
$$
 \overline{\overline{M}} \ddot{\overline{q}} =   \overline{h}_a + \overline{h}_r 
$$
dove $\overline{h}_a$ sono le forze attive mentre $\overline{h}_r$ sono le reazioni vincolari.

Nel caso di un sistem multibody con $n_b$ corpi la stessa equazione di cui sopra varrà per ogni corpo. Possiamo usare la stessa scrittura  facendo le seguenti definizioni:

**Vettore accelerazione**
$$
\ddot{\overline{q}} = \begin{bmatrix}
                    \ddot{x}_1\\
                    \ddot{y}_1\\
                    \ddot{\phi}_1\\
                    \vdots\\
                    \ddot{x}_{n_b}\\
                    \ddot{y}_{n_b}\\
                    \ddot{\phi}_{n_b}
\end{bmatrix}
$$
**Vettore forza**
$$
\overline{h} = \begin{bmatrix}
                    h_{x,1}\\
                    h_{y,1}\\
                    h_{\phi,1}\\
                    \vdots\\
                    h_{x,n_b}\\
                    h_{y,n_b}\\
                    h_{\phi,n_b}\\
\end{bmatrix}
$$

In generale il vettore delle coordinate $\overline{q}$ e delle forze $\overline{h}$ avranno dimensione $3\times n_b ,1$.

**Matrice di massa**

Per un sistem multibody la matrice delle masse sarà diagonale di dimensioni $3\times n_b ,3\times n_b$ del tipo:
$$
\overline{\overline{M}}   = \begin{bmatrix}
           m_1   & 0    & 0       & 0    & 0    & 0      & \ldots  \\
           0     & m_2  & 0       & 0    & 0    & 0      & \ldots \\
           0     & 0    & J_{G,1} & 0    & 0    & 0      & \ldots \\
           0     & 0    & 0       & m_2  & 0    & 0      & \ldots \\
           0     & 0    & 0       & 0    & m_2  & 0      & \ldots  \\
           0     & 0    & 0       & 0    & 0    & J_{G,2}& \ldots  \\
           \vdots&\vdots&\vdots   &\vdots&\vdots&\vdots  & \ddots  
         \end{bmatrix}
$$


## Relazioni del vincolo cerniera 

<img src="Lecture5_multibody_1.png" alt="Drawing" style="width: 500px;"/>
 
 Utilizziamo il caso del vincolo cerniera interna per caclolare le forze vincolari. Immaginiamo di calcolare le forze di reazione per i due corpi. Cominciamo da corpo (i):
 
 $$
 \begin{array}{rcl}
 h_{r,x}     &=& \lambda_1\\
 h_{r,y}     &=& \lambda_2\\
 h_{r,\phi} &=& -s_{P_i,y}\lambda_1 + s_{P_i,x}\lambda_2 
 \end{array}
 $$
 

dove l'ultimo terminde è l'esito del prodotto vettoriale  $\overline{s}_{P_i}\times\overline{\lambda}_{1,2}$ con $\overline{\lambda}_{1,2}$ è il vettore $\overline{\lambda}_{1,2} = [\lambda_1,\lambda_2]$. 

L'ultimo termine può essere scritto in modo equivalente come:
$$
     h_{r,\phi} = \overline{s}_{P_i,\perp}\cdot\overline{\lambda}_{1,2}
$$
Possimao scrivere in termini matriciali l'equazione come:
$$
\overline{h}_r = \begin{bmatrix} 
                  1 & 0\\
                  0 & 1\\
                  -s_{P_i,y} & s_{P_i,x}
                  \end{bmatrix}
                  \begin{bmatrix} 
                  \lambda_1\\
                  \lambda_2
                  \end{bmatrix} 
$$
dove in forma ancora più compatta:
$$
\overline{h}_r = \begin{bmatrix} 
                  \overline{\overline{I}}\\
                  \overline{s}_{P_i,\perp}^T
                  \end{bmatrix}                  
                  \begin{bmatrix} 
                  \lambda_1\\
                  \lambda_2
                  \end{bmatrix} 
$$
con $\overline{\overline{I}} = \begin{bmatrix} 1 & 0\\ 0 & 1 \end{bmatrix}$ matrice identità.

Nel caso volessimo costruire il vettore delle forze di reazione dovute al giunte P per i due corpi (i) e (j) potremmo fare:
$$
\begin{bmatrix} 
\overline{h}_{r,x,i}\\
\overline{h}_{r,y,i}\\
\overline{h}_{r,\phi,i}\\
\overline{h}_{r,x,j}\\
\overline{h}_{r,y,j}\\
\overline{h}_{r,\phi,j}\\
\end{bmatrix} 
= \begin{bmatrix} 
                  1 & 0\\
                  0 & 1\\
                  -s_{P_i,y} & +s_{P_i,x}\\
                  -1 & 0\\
                  0 & -1\\
                  +s_{P_j,y} & -s_{P_j,x}\\
                  \end{bmatrix}
                  \begin{bmatrix} 
                  \lambda_1\\
                  \lambda_2
                  \end{bmatrix} 
$$
In forma matriciale possiamo scrivere :
$$
\overline{h}_r = \begin{bmatrix} 
                  \overline{\overline{I}}\\
                  \overline{s}_{P_i,\perp}^T\\
                  -\overline{\overline{I}}\\
                  -\overline{s}_{P_j,\perp}^T
                  \end{bmatrix} 
                  \begin{bmatrix} 
                  \lambda_1\\
                  \lambda_2
                  \end{bmatrix} 
$$

## Equazione di vincolo

Il corpo vincolato oltre a soddisfare l'equazione della dinamica deve soddisfare le condizioni cinematiche che pongono i vincoli.
Supponiamo che l'equazioni di vincolo siano indicate da un sistema di di $n_c$ una per ogni vincolo del tipo:
$$
  g_1(\overline{q}) = 0\\
  g_2(\overline{q}) = 0\\
  \vdots \\
  g_{n_c}(\overline{q}) = 0
$$

Possiamo scrivere questa equazione come un equazione vettoriale:
$$
  \overline{g}(\overline{q}) = 0
$$

Le equazioni di vincolo devono rimanere invariate nel tempo e quindi la derivata diventa:
$$
    \dot{\overline{g}}(\overline{q}) = \overline{\overline{D}}\dot{\overline{q}} = 0
$$
dove $\overline{\overline{D}}$ si chiama matrice Jacobiana.

Lo stesso discorsco vale per l'accelerazione che si scrive:
$$
    \ddot{\overline{g}}(\overline{q}) = \overline{\overline{D}}\ddot{\overline{q}}  + \dot{\overline{\overline{D}}}\dot{\overline{q}} = 0
$$
l'ultima equazione si può scrivere nel seguente modo:
$$
    \overline{\overline{D}}\ddot{\overline{q}}  = - \dot{\overline{\overline{D}}}\dot{\overline{q}} = \overline{\gamma}
$$
dove $\overline{\gamma}$ è un termine quadratoco di velocità.

 ## Equazioni del vincolo cerniera
 
 <img src="Lecture5_multibody_2.png" alt="Drawing" style="width: 500px;"/>
 
 L'equazione di vincolo si può parafrasare richiedendo che la coordinata x e y del punto $P_i$ e $P_j$ sono uguali. Si possono scrivere 2 equazioni che in forma vettoriale diventano:
 
 $$
     \overline{g}(\overline{q}) = \overline{r}_{P_i} -\overline{r}_{P_j} = 0
 $$
 
 oppure:
 
 $$
     \overline{g}(\overline{q}) = \overline{r}_{G_i}  +\overline{\overline{R}}(\phi_i)\overline{\sigma}_{P_i}  -\overline{r}_{G_j}  -\overline{\overline{R}}(\phi_j)\overline{\sigma}_{P_j}
 $$
 L'equazione della velocità si scrivono come:
 
 $$
      \dot{\overline{g}}(\overline{q}) =  \dot{\overline{r}}_{P_i} -  \dot{\overline{r}}_{P_j} = 0
 $$
 
 Questo si può scrivere come:
 
 $$
  \dot{\overline{g}}(\overline{q}) =  \dot{\overline{r}}_{G_i}  + \dot{\phi}_i \overline{s}_{P_i,\perp}  -\dot{\overline{r}}_{G_j}  - \dot{\phi}_j \overline{s}_{P_j,\perp} = 0
 $$
 in forma matriciale possiamo scrivere:
 
$$
  \begin{bmatrix}
  1 &  0 & -s_{P_i,y} & -1 &  0 & +s_{P_j,y}\\
  0 &  1 & s_{P_i,x}  &  0 & -1 & -s_{P_j,x}
  \end{bmatrix}
     \begin{bmatrix}
   \dot{x}_i\\
   \dot{y}_i\\
   \dot{\phi}_i\\
   \dot{x}_j\\
   \dot{y}_j\\
   \dot{\phi}_j\\
   \end{bmatrix} = 0
$$
 ed in forma compatta diventa:
$$
   \begin{bmatrix}
   \overline{\overline{I}} & \overline{s}_{P_i,\perp} & -\overline{\overline{I}} & -\overline{s}_{P_j,\perp}
  \end{bmatrix}
     \begin{bmatrix}
   \dot{x}_i\\
   \dot{y}_i\\
   \dot{\phi}_i\\
   \dot{x}_j\\
   \dot{y}_j\\
   \dot{\phi}_j\\
   \end{bmatrix} = 0
$$
   Per riferimento possiamo chiamare le matrici $\overline{{D}}_i$ e $\overline{{D}}_j$ come:
   
$$
   \overline{\overline{D}}_i   = \begin{bmatrix}
   \overline{\overline{I}} & \overline{s}_{P_i,\perp} 
  \end{bmatrix}\quad \text{e}\quad    \overline{\overline{D}}_j   = \begin{bmatrix}
   -\overline{\overline{I}} & -\overline{s}_{P_j,\perp} 
  \end{bmatrix}
$$
    

L'accelerazione si può scrivere come:
$$
     \ddot{\overline{g}}(\overline{q}) =     \begin{bmatrix}
   \overline{\overline{I}} & \overline{s}_{P_i,\perp} & -\overline{\overline{I}} & -\overline{s}_{P_j,\perp}
  \end{bmatrix}    
  \begin{bmatrix}
   \ddot{x}_i\\
   \ddot{y}_i\\
   \ddot{\phi}_i\\
   \ddot{x}_j\\
   \ddot{y}_j\\
   \ddot{\phi}_j\\
   \end{bmatrix} 
   +
    \begin{bmatrix}
   \overline{\overline{0}} & \dot{\overline{s}}_{P_i,\perp} & -\overline{\overline{0}} & -\dot{\overline{s}}_{P_j,\perp}
  \end{bmatrix}  
   \begin{bmatrix}
   \dot{x}_i\\
   \dot{y}_i\\
   \dot{\phi}_i\\
   \dot{x}_j\\
   \dot{y}_j\\
   \dot{\phi}_j\\
   \end{bmatrix} = 0
    $$
Per riferimento possiamo chiamare le matrici $\overline{{D}}_i$ e $\overline{{D}}_j$ come:
   
   $$
   \dot{\overline{\overline{D}}}_i   = \begin{bmatrix}
   \overline{\overline{0}} & \dot{\overline{s}}_{P_i,\perp} 
  \end{bmatrix}\quad \text{e} \quad  \dot{\overline{\overline{D}}}_j   = \begin{bmatrix}
   -\overline{\overline{0}} & -\dot{\overline{s}}_{P_j,\perp} 
  \end{bmatrix}
   $$
con $\overline{\overline{0}} = \begin{bmatrix} 0 & 0\\ 0 & 0 \end{bmatrix}$ matrice nulla.

Ricordiamo che $\dot{\overline{s}}_{P_i,\perp}   =  \overline{\dot{s}}_{P_i,\perp}$.

In definitiva 
$$
     \ddot{\overline{g}}(\overline{q}) =     \begin{bmatrix}
   \overline{{D}}_i & \overline{{D}}_j
  \end{bmatrix}    
  \begin{bmatrix}
   \ddot{x}_i\\
   \ddot{y}_i\\
   \ddot{\phi}_i\\
   \ddot{x}_j\\
   \ddot{y}_j\\
   \ddot{\phi}_j\\
   \end{bmatrix} 
   +
   \dot{\overline{s}}_{P_i,\perp} \dot{\phi}_i - \dot{\overline{s}}_{P_j,\perp} \dot{\phi}_j = 0
$$

## Realazione tra reazioni vincolari e vettore delle forze per una cerniera 

Si può notare i vettore delle forze per la cerniera si possa scrivere come:
$$
\overline{h}_r = \overline{\overline{D^T }}\overline{\lambda}_{1,2}
$$
dove la matrice $\overline{\overline{D^T }}$ è la trasposta di quella definita nella sezione precedente.
Dal punto di vista teorico questa relazione dipende dal fatto che le forze di reazione possono hanno "direzione" sempre normale alle superficie descritte dalla funzioni di vincolo.

In generale possiamo scrivere:
$$
\overline{h}_r = \overline{\overline{D^T }}\overline{\lambda}
$$

## Equazioni del moto per un corpo vincolato
Mettendo assieme tutte le relazioni che si sono trovate possiamo scrivere in generale:
$$
\begin{array}{rcl}
 \overline{\overline{M}} \ddot{\overline{q}} &=&   \overline{h}_a + \overline{\overline{D^T }}\overline{\lambda}\\
  \overline{\overline{D}}\ddot{\overline{q}} &=& - \dot{\overline{{D}}}\dot{\overline{q}}
  \end{array}
$$
 
 evidenziamo che le incognite del problema sono le accelerazioni $\ddot{\overline{q}}$ e le le reazioni vincolare $\overline{\lambda}$.
 Per disaccoppiare le due equazioni si può riscrivere il sistema nel seguente modo:
 $$
\begin{array}{rcl}
 \ddot{\overline{q}} &=&   \overline{\overline{M}}^{-1}(\overline{h}_a + \overline{\overline{D^T }}\overline{\lambda})\\
  \overline{\overline{D}}\overline{\overline{M}}^{-1}(\overline{h}_a + \overline{\overline{D^T }}\overline{\lambda}) &=& - \dot{\overline{{D}}}\dot{\overline{q}}
  \end{array}
$$
che diventa:
 $$
\begin{array}{rcl}
 \ddot{\overline{q}} &=&   \overline{\overline{M}}^{-1}(\overline{h}_a + \overline{\overline{D^T }}\overline{\lambda})\\
  \overline{\overline{D}}\overline{\overline{M}}^{-1} \overline{\overline{D^T }}\overline{\lambda} &=& - \dot{\overline{{D}}}\dot{\overline{q}} -\overline{\overline{D}}\overline{\overline{M}}^{-1}\overline{h}_a
  \end{array}
$$
E' interessante notare che la seconda equazione può essere calcolata ogni qual volta si conoscano le posizione e le velocità delle coordinate del problema, questo ci suggerisce che una possibile soluzione al problema dell'integrazione numerica del problema multibody potrebbe essere:

**Integrazione temporale**

0. ricavare i valori al tempo i-esimo della posizione, $\overline{q}_i$ , e della velocità   $\dot{\overline{q}}_i$ 
1. calcolare le reazioni vincolari al tempo i-esimo utilizzando la seconda equazione del sistema
$$
 \overline{\lambda} = -(\overline{\overline{D}}\overline{\overline{M}}^{-1} \overline{\overline{D^T }})^{-1} ( \dot{\overline{{D}}}\dot{\overline{q}} + \overline{\overline{D}}\overline{\overline{M}}^{-1}\overline{h}_a)
$$
2. calcolare le accelerazioni all'instante i-esimo $\ddot{\overline{q}}_i$ e integrare le equazioni del moto

## Violazione dei vincoli e metodo della penalità

La soluzione numerica delle equazioni della dinamica porta inevitabilmente ad errori che si traducono nella violazione delle condizion di vincolo.

Per ovviare a questo problema si può considerare di sostituire i vincoli e le forze vincolari scambiate con una combinazioni di molle/smorzatori fittizi che prendano direttamente in conto la violazione dei vincoli.
Partendo dalla:

$$
\overline{\overline{M}} \overline{q} =   \overline{h}_a + \overline{\overline{D^T}} \overline{\lambda} 
$$

facciamo la seguente ipotesi:
$$
\overline{\lambda}  =-( m_p \ddot{\overline{g}} + b_p \dot{\overline{g}} +  k_p \overline{g})
$$
dove ora le reazioni vincolari diventano fittizzie e sono considerate come forze che dipendono dalla violazione globale dei vincoli e hanno la forma di molle di constante $k_p$ smorzatori di constante $b_p$ o masse aggiunte di constante $m_p$.

<img src="Lecture5_multibody_4.png" alt="Drawing" style="width: 500px;"/>

L'equazione della dinamica sarà:
$$
\overline{\overline{M}} \overline{q} =   \overline{h}_a - \overline{\overline{D^T}}( m_p \ddot{\overline{g}} + b_p \dot{\overline{g}} +  k_p \overline{g})
$$

esprimendo la relazione che c'è tra la derivata seconda della funzioni di vincolo e le coordinate $\ddot{\overline{g}}(\overline{q}) = \overline{\overline{D}}\ddot{\overline{q}}  + \dot{\overline{\overline{D}}}\dot{\overline{q}}$ e portando le accelerazioni tutte da una parte possiamo scrivere: 

$$
(\overline{\overline{M}}  + m_p \overline{\overline{D^T}}\overline{\overline{D}}\ddot{\overline{q}}) =   \overline{h}_a - \overline{\overline{D^T}}( m_p \dot{\overline{\overline{D}}}\dot{\overline{q}} + b_p \dot{\overline{g}} +  k_p \overline{g})
$$

Una forma equivalente è la seguente:
$$
(\overline{\overline{M}}  + m_p \overline{\overline{D^T}}\overline{\overline{D}}\ddot{\overline{q}}) =   \overline{h}_a -m_p  \overline{\overline{D^T}}( \dot{\overline{\overline{D}}}\dot{\overline{q}} + \frac{b_p}{m_p} \dot{\overline{g}} +  \frac{k_p}{m_p} \overline{g})
$$

Mimando il sistema massa molla e definendeo le seguenti quantità fittizie: frequenza naturale $\omega = \sqrt {\frac{k_b}{m_b}}$ e smorzamento  $\zeta = \frac{b_p}{2\sqrt{k_p m_p}}$ possiamo riscrivere l'equazione come:

$$
(\overline{\overline{M}}  + m_p \overline{\overline{D^T}}\overline{\overline{D}})\ddot{\overline{q}} =   \overline{h}_a -m_p  \overline{\overline{D^T}}( \dot{\overline{\overline{D}}}\dot{\overline{q}} + 2\zeta\omega \dot{\overline{g}} +  \omega^2 \overline{g})
$$


Si noti come in questo modo spariscano le incognite legate alle forze dei vincoli ed il problema possa essere risolto molto più semplice.

**Integrazione temporale**

0. ricavare i valori al tempo i-esimo della posizione, $\overline{q}_i$ , e della velocità   $\dot{\overline{q}}_i$ 
1. calcolare le accelerazioni all'instante i-esimo $\ddot{\overline{q}}_i$ e integrare le equazioni del moto
$$
\ddot{\overline{q}} =   (\overline{\overline{M}}  + m_p \overline{\overline{D^T}}\overline{\overline{D}})^{-1}\Bigl(\overline{h}_a -m_p  \overline{\overline{D^T}}\bigl( \dot{\overline{\overline{D}}}\dot{\overline{q}} + 2\zeta\omega \dot{\overline{g}} +  \omega^2 \overline{g}\bigl)\Bigl)
$$

**Note**

Bisogna fare però delle considerazioni sui valori di $m_p$,$b_p$ e $k_p$. Per limitare le violazioni dei vincoli di posizione bisogna avere una rigidezza elevata $k_p$ e sicuramente più grande della rigidezze tipiche del problema. Avere una rigidezza elevata significa avere potenzialmente un sistema con un alta frequenze naturali di oscillazione $\omega$ che può generare disturbi nella soluzione in alta frequenza. Per abbasare la frequenza dei disturbi bisogna ammettere quindi termini di massa fittizzia elevati $m_p$ di solito si prendeono valori almeno 3 ordini di grandezza le masse tipiche del sistema. Definita la massa si assume un valore di termine di smorzamento pari a uno $\zeta = 1$. Inotre si stabilisce il valore della frequenza fittizia in modo tale che il perido di oscillazione risulti $T = \frac{2\pi}{\omega} = 0.1 - 1 s$. Chiaramente il disturbo dovuto al sistema fittizzio di massa molla dovrebbe essere lontano dalle frequenze naturali proprie del sistema. 


## Pendolo fisico
 <img src="Lecture5_multibody_3.png" alt="Drawing" style="width: 200px;"/>
Il pendolo fisico è un rigido sospeso ad un punto A e soggetto alla forza di gravità. L'equazione per il solo grado di libertà della rotazione introno al punto A come:

$$
J_A \ddot{\theta } = \mu_A 
$$

Il momento della forza sara:  $\mu_A = - mgl\sin(\theta)$ dove m è la massa del pendolo e  $l$ è la distanza del baricentro dal polo.

L' equazione differenziale non lineare associatà sarà:

$$
\ddot{\theta }  + \frac{mgl}{J_A}\sin(\theta) = 0
$$

dove assumiamo che valga $\theta(t=0) = 0$ e assumiamo il massimo valore di $\theta$ uguale a $\theta_{max}$.

Nel caso del punto di massa $J_A$ diventa $ J_A = ml^2$ e quindi si ha l'equazione del pendolo tradizionale:

$$
\ddot{\theta}  + \frac{g}{l}\sin(\theta) = 0
$$

Nel caso le oscillazioni siano piccole ,$\theta_{max}<<1$, l'equazione si approssima:

$$
\ddot{\theta }  + \frac{mgl}{J_A}\theta = 0
$$

Definendo poi $\omega = \sqrt{\frac{mgl}{J_A}}$ la soluzione diventa:

$$
\theta(t) = \theta_{max} \sin(\omega t)
$$

Questa soluzione può ritenersi valida per oscillazioni inferiori a $\theta_{max}<10°$.

Nel caso del pendolo non lineare la soluzione diventa molto differente si ha in genere che la frequenza di oscillazione dipende dall'ampiezza massima e la soluzione non è più perfettamente armonica.

La soluzione a cui si perviene si può scrivere come:

$$
\theta(t) = 2 \arcsin \bigl( k \cdot \text{sn}\left(\omega t;k\right) \bigl) \quad \text{con} \quad k = sin(\theta_{max}/2)
$$

dove la funzione $\text{sn}(\omega t;k)$ è la funzione sn elittica di Jacobi ,si veda https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.ellipj.html.

Soluzione pag 111 Fasano Marmi Meccanica Analitica.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import ellipj

w = 2
T = 2*np.pi/w
nper = 2
t = np.linspace(0,nper*T,200)

theta_max = np.pi/180*1# °1 deg


theta_lin  = theta_max*np.sin(w*t) # linear angle
dtheta_lin = w*theta_max*np.cos(w*t) # linear angular velocity

k = np.sin(theta_max/2)
sncndnph = ellipj(w*t,k) #output sn cn dn ph
sn = sncndnph[0]; # sine amplitude
cn = sncndnph[1]; # cosine amplitude

theta_nonlin  = 2*np.arcsin(k*sn) # nonlinear angle
dtheta_nonlin = 2*k*w*cn # nonlinear angular velocity

fig, axs = plt.subplot_mosaic([['traj', 'pos'],['traj', 'vel']])

axs['pos'].plot(t,theta_lin,label  = 'linear')
axs['pos'].plot(t,theta_nonlin,label  = 'nonlinear')
axs['pos'].set_title(" posizione")

axs['vel'].plot(t,dtheta_lin,label  = 'linear')
axs['vel'].plot(t,dtheta_nonlin,label  = 'nonlinear')
axs['vel'].set_title(" velocità")

axs['traj'].plot(theta_lin,dtheta_lin,label  = 'linear')
axs['traj'].plot(theta_nonlin,dtheta_nonlin,label  = 'nonlinear')
axs['traj'].set_title(" spazio delle fasi")
axs['traj'].set_xlabel(r"$\theta$")
axs['traj'].set_ylabel(r"$\dot{\theta}$")
axs['traj'].legend()
axs['traj'].set_aspect('equal')
```
