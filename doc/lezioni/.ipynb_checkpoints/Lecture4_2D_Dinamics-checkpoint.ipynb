# Cinematica e dinamica di un corpo rigido in due dimensioni
## Introduzione

Nel caso di un corpo rigido libero sottoposto ad una serie di forze e momenti le equazioni della dinamica rispetto ad un riferimento fisso inerziale si posso scrivere nel seguente modo:

$$
  \begin{eqnarray}
  m\ddot{x_G} &=& f_x\\
  m\ddot{y_G} &=& f_y\\
  J_G\ddot{\phi} &=& \mu_G
  \end{eqnarray}
$$

dove $\ddot{x_G}$ e $\ddot{y_G}$ sono le accelerazioni del baricentro e $\ddot{\phi}$ è l'accelerazione angolare del corpo rispetto ad un sistema di riferimento fisso e dove $J_G$ è il momento di inerzia assiale di massa rispetto al baricentro e $\mu_G$ è il momento risultante delle forze rispetto al baricentro.

Nel caso generale $\mu_G$ si può calcolare come:
$$
    \mu_G = \sum_i^N \mu_i + \sum_j^M \overline{s}_{A_j,\perp}\cdot \overline{f}_j 
$$

dove $\mu$ son le coppie pure e $s_{A_j,\perp}\cdot f_j$ sono i momenti delle forze applicate in $A_j$.

Dal punto di vista matriciale si può scrivere equivalentemente:

$$
\begin{bmatrix}
           m & 0 & 0\\
           0 & m &0\\
           0 &  0 & J_G 
\end{bmatrix}
\begin{bmatrix}
           \ddot{x_G}\\
           \ddot{y_G}\\
           \ddot{\phi} 
\end{bmatrix} = 
\begin{bmatrix}
           f_x\\
           f_y\\
           \mu_G 
\end{bmatrix}
$$

Più in generale possiamo scrivere
$$
 \overline{\overline{M}} \ddot{\overline{q}} =  \overline{h}
 $$
 dove $\overline{\overline{M}}$ è detta matrice di massa 
 $$
 \overline{\overline{M}}=
 \begin{bmatrix}
           m & 0 & 0\\
           0 & m &0\\
           0 &  0 & J_G 
\end{bmatrix}
 $$
 
 $\ddot{\overline{q}}$ è il vettore delle accelerazioni 
 
 $$
 \ddot{\overline{q}} = \begin{bmatrix}
           \ddot{x_G}\\
           \ddot{y_G}\\
           \ddot{\phi} 
\end{bmatrix}
 $$
 
 e  $\overline{h}$ è il vettore di tutte le forze esterne:
 $$
 \overline{h}  = \begin{bmatrix}
           f_x\\
           f_y\\
           \mu_G 
\end{bmatrix}
 $$


## Cinematica di un rigido

**Posizione**

<img src="Lecture4_reference_frames.png" alt="Drawing" style="width: 500px;"/>

Sia dato un sistema di riferimento fisso O,x,y e un sistema di riferimento mobile e solidale al corpo centrato nel baricentro G,$\xi,\eta$. Sia inoltre il vettore $\overline{s}_A$ il vettore ottenuto ruotando il vettore nella terna solidale, G,$\xi,\eta$ ,$\overline{\sigma}_A$  nella posizione corrente e cioè $ \overline{s}_A =\overline{\overline{R}}\overline{\sigma}_A$. Allora  il vettore posizione $\overline{r}_A$ si può scrivere come:
$$
\overline{r}_A =\overline{r}_G + \overline{s}_A =  \overline{r}_G  + \overline{\overline{R}}\overline{\sigma}_A
$$

dove $\overline{\overline{R}}$ è la matrice di rotazione di angolo $\phi$ dove

$$
\overline{\overline{R}}(\phi) = \begin{bmatrix}
           \cos(\phi)& -\sin(\phi) \\
           \sin(\phi)& \cos(\phi)
       \end{bmatrix} 
$$

**Velocità**

Per quanto rigurda la velocità del punto A si può scirivere come:
$$
\dot{\overline{r}}_A = \dot{\overline{r}}_G  + \dot{\overline{s}}_A = \dot{\overline{r}}_G + \dot{\overline{\overline{R}}}\overline{\sigma}_A
$$

<img src="Lecture4_2D_dynamics.png" alt="Drawing" style="width: 500px;"/>

La derivata della matrice di rotazione si può scrivere come:
$$
\dot{\overline{\overline{R}}}(\phi)=\dot{\phi}\begin{bmatrix}
           -\sin(\phi)& -\cos(\phi) \\
           \cos(\phi)& -\sin(\phi)
\end{bmatrix} 
$$

ora dalle relazioni si può notare $\cos(\phi + \pi/2) = -\sin(\phi)$ e $\sin(\phi + \pi/2) = \cos(\phi)$

quindi possiamo riscrivere la matrice $\dot{\overline{\overline{R}}}(\phi)$ come:
$$ \dot{\overline{\overline{R}}}(\phi)  =  \dot{\phi}\overline{\overline{R}}(\phi + \pi/2)$$

Ora analizzando il secondo termine della velocità ci accorgiamo che :
$$
\dot{\overline{s}}_A  = \dot{\overline{\overline{R}}}\overline{\sigma}_A  =  \dot{\phi}\overline{\overline{R}}(\phi + \pi/2)\overline{\sigma}_A =\dot{\phi}\overline{s}_{A,\perp}   
$$


dove $\overline{\overline{R}}(\phi + \pi/2)\overline{s}_A = \overline{s}_{A,\perp}$ è il vettore ortogonale al vettore $s_A$ con lo stesso modulo.

Si noti che il vettore $\dot{\overline{s}}_A$ è un vettore ortogonale routato di +$\pi/2$ rispetto al vettore posizione ${\overline{s}}_A$.

Inoltre è molto semplice costruire il vettore $\overline{s}_{A,\perp}$ infatti:
$$
\overline{s}_{A,\perp} = [-s_{A,y},s_{A,x}]^T
$$

Quindi possiamo scrivere:
$$
\dot{\overline{r}}_A = \dot{\overline{r}}_G  + \dot{\phi}\overline{s}_{A,\perp} 
$$


## Esercizio
0. provare a costruire i vettore normale ad un vettore dato


## Accelerazione

Calcoliamo ora l'accelerazione del punto A. 
$$
\ddot{\overline{r}}_A = \ddot{\overline{r}}_G  + \ddot{\overline{s}}_A = \ddot{\overline{r}}_G + \ddot{\overline{\overline{R}}}\overline{\sigma}_A
$$
Ma la derivata seconda della matrice di rotazione sarà
$$
\ddot{\overline{\overline{R}}}(\phi)=
\ddot{\phi}\underbrace{\begin{bmatrix}
           -\sin(\phi)& -\cos(\phi) \\
           \cos(\phi)& -\sin(\phi)
\end{bmatrix}}_{\overline{\overline{R}}(\phi+\pi/2)} - 
\dot{\phi}^2\underbrace{\begin{bmatrix}
           \cos(\phi)& -\sin(\phi) \\
           -\sin(\phi)& \cos(\phi)
\end{bmatrix}}_{\overline{\overline{R}}(\phi)}
$$

Quindi possiamo scrivere:

$$
\ddot{\overline{s}}_A = \ddot{\overline{\overline{R}}}\overline{\sigma}_A  = \ddot{\phi}\overline{s}_{A,\perp}  - \dot{\phi}^2\overline{s}_{A}  
$$

Quindi il vettore  $\ddot{\overline{s}}_A$  avrà due componenti una tangente alla traiettoria di rotazine del punto A,  $\ddot{\phi}\overline{s}_{A,\perp} $ e una lungo la congintente il punto A ed il baricentro G  e diretta verso il baricentro $- \dot{\phi}^2\overline{s}_{A}$.

E quindi:

$$
\ddot{\overline{r}}_A = \ddot{\overline{r}}_G  + \ddot{\phi}\overline{s}_{A,\perp}  - \dot{\phi}^2\overline{s}_{A} 
$$

## Forza di una molla
Assumiamo la presenza di una molla fissata a due punti A e B eventualmente in moto relativo appertenenti a due corpi diversi (i) e (j).

<img src="Lecture4_spring_damper.png" alt="Drawing" style="width: 500px;"/>

La componente lungo la direzione indicata da versore $\hat{d}$ della forza della molla è legata all'allungamento e cioè a $\Delta L = L-L_0$ dove $L_0$ è la lunghezza della molla a riposo e alla costante elastica come:
$$
    f_k = k(L-L_0)
$$

Il vettore forza si può quindi scrivere per il corpo (i) come

$$
    \overline{f}_{k,(i)} =   - \hat {d} f_k
$$
E per il corpo (j)

$$
    \overline{f}_{k,(j)} =   + \hat {d} f_k
$$

Definiamo il versore $\hat{d}$ come  $ \hat{d} = \frac{\overline{d}}{|\overline{d}|} $ dove
$$
\overline{d} = \overline{r}_A - \overline{r}_B 
$$

Notiamo inoltre che la lunghezza L concide con il modulo del vettore $\overline{d}$,

$$
L  =|\overline{d}|
$$

## Forza di uno smorzatore

Assumiamo la presenza di uno smorzatore fissato a due punti A e B eventualmente in moto relativo appertenenti a due corpi diversi (i) e (j).

La componente lungo la direzione indicata da versore $\hat{d}$ della forza della molla è legata alla velocità di allungamento e cioè a $\dot{\Delta L} = \dot{L}$ come:

$$
    f_b = b\dot{L}
$$
Ora possiamo calcolare $\dot{L}$ come 

$$
 \dot{L} = \hat{d}\cdot\dot{\overline{d}}
$$
dove si intende il prodotto scalare e dove in questo caso:
$$
\dot{\overline{d}} = \dot{\overline{r}}_A - \dot{\overline{r}}_B 
$$

Il vettore forza si può quindi scrivere per il corpo (i) come

$$
    \overline{f}_{b,(i)} =   - \hat {d} f_b
$$
E per il corpo (j)

$$
    \overline{f}_{b,(j)} =   + \hat {d} f_b
$$

## Esercizi 
1. Corpo con molla posizionata nel baricentro e con posizione iniziale posizioni $x_0$,$y_0$ e $\phi_0$ no gravità
2. Corpo con molla e smorzatore posizionata nel baricentro e con posizione iniziale posizioni $x_0$,$y_0$ e $\phi_0$ no gravità 
3. Corpo con molla e smorzatore posizionata nel baricentro e con posizione iniziale posizioni $x_0$,$y_0$ e $\phi_0$ con la gravità
4. Corpo con molla posizionata in un punto generico A e con posizione iniziale posizioni $x_0$,$y_0$ e $\phi_0$ no gravità
